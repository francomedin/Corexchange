```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.style
plt.style.use('seaborn')
```

```python
data=pd.read_csv('/content/drive/MyDrive/Datos Colab/TSLA.csv')
```

```python
#Convierto de string a datetime
data['Date']=pd.to_datetime(data['Date'])
```

```python
#Convierno la columna Date a indice
data=data.set_index('Date')
```

```python
data['Close'].plot()
```

```python
data['Close'].plot(figsize=(50,20),lw=3)
```

```python
graf_sup_izquierda=plt.subplot2grid(shape=(4,1),loc=(0,0),rowspan=3,colspan=4)
graf_sup_izquierda.plot(data.index,data['High'])
graf_sup_izquierda.set_ylabel('Maximo')

#graf_sup_derecho=plt.subplot2grid(shape=(4,4),loc=(0,2),rowspan=2,colspan=2)
#graf_sup_derecho.bar(data.index,data['Low'])
#graf_sup_derecho.set_xlabel('Minimo')

graf_inferior=plt.subplot2grid(shape=(4,1),loc=(3,0),rowspan=1,colspan=4)
graf_inferior.bar(data.index,data['Volume'])
graf_inferior.set_xlabel('Volumen')

plt.gcf().set_size_inches(15,8)
```

```python

graf_superior=plt.subplot2grid(shape=(4,4),loc=(0,0),rowspan=2,colspan=4)
graf_superior.plot(data['Close'])
graf_superior.set_xlabel('Cierre')

graf_inferior=plt.subplot2grid(shape=(4,4),loc=(3,0),rowspan=2,colspan=4)
graf_inferior.plot(data['Volume'])
plt.gcf().set_size_inches(15,8)
graf_inferior.set_xlabel('Volumen')
```
